```python
import pandas as pd
import folium

d = pd.read_excel('전국도시공원정보표준데이터-20220927.xls')
d=d.dropna(subset=['소재지지번주소','위도','경도'])
d=d.query('소재지지번주소.str.contains("서울")', engine = 'python')

m=folium.Map(location=[37,127])

for i in range(len(d)):
    folium.Marker([float(d['위도'][i:i+1]), float(d['경도'][i:i+1])]).add_to(m)
    
m
```
