```python
import wordcloud
import matplotlib.pyplot as plt

text="Soongsil univ Ai Ai BigData Soongsil Soongsil Ai Logo Logo Logo" #, 횟수가 많을 수록 텍스트 크게 출현

wc= wordcloud.WordCloud(width=1500,height=1000).generate(text)

plt.figure(figsize=(5,2)) #전체크기
plt.axis('off') #축 제거
plt.imshow(wc)# 이미지로 출력
plt.show()
```

```python
import wordcloud
from wordcloud import STOPWORDS

u_word=wordcloud.STOPWORDS.union({'AI'})#STOPWORD:불용어, union:합집합, AI 제거

wc= wordcloud.WordCloud(width=1500,height=1000,stopwords=u_word).generate(text)


import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.axis('off')
plt.imshow(wc)
plt.show()

wc.to_file('a.jpg')
#files.download('a.jpg')
```

```python
from  wordcloud import WordCloud,STOPWORDS #wordcloud는 코드 쓰는 시험은 없음
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np


plt.rc('font',family='Gulim') # 한글폰트 사용
s_words=STOPWORDS.union(['같은','많은','있는', '이', '대한']) # 불용어 및 지정단어 합집합하여 데이터에서 제거
maskheart = np.array(Image.open('figure.jpg'))

wordcloud= WordCloud(width=1500,height=1000,stopwords=s_words,mask=maskheart, 
                     background_color='white',font_path='c:/windows/fonts/Gulim.ttc').generate(text)

plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```
